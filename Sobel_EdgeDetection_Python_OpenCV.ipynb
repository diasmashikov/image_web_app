```
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
print('OpenCV Version: ', cv2.__version__)
```

```
original_img_1 = cv2.imread('/content/drive/My Drive/Tutorials/Data/boyd.png')
h, w, c = original_img_1.shape

original_img_1_height, original_img_1_width = original_img_1.shape[:2]
print('Original Image #1 Width: ', original_img_1_width)
print ('Original Image #1 Height: ', original_img_1_height)

plt.figure(figsize = (10,10))

plt.subplot(121),plt.imshow(cv2.cvtColor(original_img_1, cv2.COLOR_BGR2RGB)),plt.title('Original Image #1')
plt.xticks([]), plt.yticks([])
plt.show()
```

```
# Edge detection using Sobel algorithms 

# Sobel edge detector for RGB image
sobelEdgesColor_x = cv2.Sobel(original_img_1,cv2.CV_64F,1,0,ksize=5)  # Vertical Edges
sobelEdgesColor_y = cv2.Sobel(original_img_1,cv2.CV_64F,0,1,ksize=5)  # Horizontal Edges

# Sobel edge detector for Gray image
original_img_1_gray = cv2.cvtColor(original_img_1, cv2.COLOR_BGR2GRAY)
sobelEdgesGray_x = cv2.Sobel(original_img_1_gray,cv2.CV_64F,1,0,ksize=5)  # Vertical Edges
sobelEdgesGray_y = cv2.Sobel(original_img_1_gray,cv2.CV_64F,0,1,ksize=5)  # Horizontal Edges

plt.figure(figsize = (20,20))

plt.subplot(321),plt.imshow(cv2.cvtColor(original_img_1, cv2.COLOR_BGR2RGB)),plt.title('Original Image')
plt.xticks([]), plt.yticks([])
plt.subplot(323),plt.imshow(sobelEdgesColor_x),plt.title('Sobel Result - Color - Vertical')
plt.xticks([]), plt.yticks([])
plt.subplot(324),plt.imshow(sobelEdgesColor_y),plt.title('Sobel Result - Color - Horizontal')
plt.xticks([]), plt.yticks([])
plt.subplot(325),plt.imshow(sobelEdgesGray_x, cmap = 'gray'),plt.title('Sobel Result - Gray - Vertical')
plt.xticks([]), plt.yticks([])
plt.subplot(326),plt.imshow(sobelEdgesGray_y, cmap = 'gray'),plt.title('Sobel Result - Gray - Horizontal')
plt.xticks([]), plt.yticks([])

plt.show()
```
